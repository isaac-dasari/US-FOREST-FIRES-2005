```python
#Helper for pretty formatting for Spark DataFrames
def showDF(df, limitRows =  20, truncate = True):
    if(truncate):
        pd.set_option('display.max_colwidth', 50)
    else:
        pd.set_option('display.max_colwidth', 1)
    pd.set_option('display.max_rows', limitRows)
    display(df.limit(limitRows).toPandas())
    pd.reset_option('display.max_rows')
```

```python
def create_spark_session(appname):
    """Creates SparkSession
        
    Creates SparkSession and returns it. If SparkSession is already created it returns
    the currently running SparkSession.
    
    Args:
        App Name
    Returns:
        SparkSession
    """
    spark = SparkSession.builder \
    .appName(appname)\
    .getOrCreate()
   
    return spark
```

```python
def visualize_missing_values_spark(df):
    """Visualize missing values in a spark dataframe
    
    :param df: spark dataframe
    """
    # create a dataframe with missing values count per column
    nan_count_df = df.select([count(when(isnan(c) | col(c).isNull(), c)).alias(c) for c in df.columns]).toPandas()
    
    # convert dataframe from wide format to long format
    nan_count_df = pd.melt(nan_count_df, var_name='cols', value_name='values')
    
    # count total records in df
    total = df.count()
    
    # now lets add % missing values column
    nan_count_df['% missing values'] = 100*nan_count_df['values']/total
    
    plt.rcdefaults()
    plt.figure(figsize=(10,5))
    ax = sns.barplot(x="cols", y="% missing values", data=nan_count_df)
    ax.set_ylim(0, 100)
    ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
    plt.show()
```

```python
visualize_missing_values_spark(firesDF)
```

# Dynamically Get columns with input percent

```python
def percent_missing_coulumns(df,percent):
    # create a dataframe with missing values count per column
    a=df.select([count(when(isnan(c) | col(c).isNull(), c)).alias(c) for c in df.columns]).toPandas()
    
    # convert dataframe from wide format to long format
    count_df = pd.melt(a, var_name='cols', value_name='values')
    
    # count total records in df
    total = df.count()
    
    # now lets add % missing values column
    count_df['%missingvalues'] = 100*count_df['values']/total
    count_df = count_df[count_df['%missingvalues']>=percent]
    return count_df
```

Get list of Columns with > 90% nulls

```python
drop_columns = percent_missing_coulumns(firesDF,90)['cols']
```

```python
#drop columns with more than 90% values dynamically
firesDF = firesDF.drop(*drop_columns)
# drop rows where all elements are missing
firesDF = firesDF.dropna(how='all')
#Drop any duplicates if exists
firesDF=firesDF.dropDuplicates()
firesDF.count()
```

# Build Dim/Fact Tables

```python
#Get Distinct Fire Name, Code, Year for Dimensational Fires table
DimFiresDF=firesDF.select('FIRE_YEAR','FIRE_CODE','FIRE_NAME').distinct()
#Generate Key Column
DimFiresDF=DimFiresDF.withColumn('FIRE_KEY',F.row_number().over(Window.orderBy(F.col("FIRE_NAME").asc()))+100000)
```

```python
DimFiresDF.select('FIRE_KEY','FIRE_YEAR','FIRE_CODE','FIRE_NAME').distinct().show(2)
```

```python
#Build FactFires Data Frame and join the fire key from DimFires
factfiresDF=firesDF.alias('a').join(DimFiresDF.alias('b'),(firesDF.FIRE_NAME==DimFiresDF.FIRE_NAME) & (firesDF.FIRE_CODE==DimFiresDF.FIRE_CODE) \
             & (firesDF.FIRE_YEAR==DimFiresDF.FIRE_YEAR),'inner').select('b.FIRE_KEY','a.*')
```

```python
visualize_missing_values_spark(factfiresDF)
```

```python
#Create Date Keys for Discovery Date and Contained Date
factfiresDF=factfiresDF.withColumn('DISCOVERY_DATE_KEY', regexp_replace('a.DISCOVERY_DATE', '-', ''))
factfiresDF=factfiresDF.withColumn('CONT_DATE_KEY', regexp_replace('a.CONT_DATE', '-', ''))
factfiresDF.select('a.DISCOVERY_DATE','DISCOVERY_DATE_KEY','CONT_DATE_KEY','CONT_DATE').show(5)
```

```python
#Create Dim Date Data Frame
DimDatesDF=factfiresDF.select('DISCOVERY_DATE_KEY','a.DISCOVERY_DATE','DISCOVERY_DOY').distinct()
DimDatesDF=DimDatesDF.union(factfiresDF.select('CONT_DATE_KEY','a.CONT_DATE','CONT_DOY').distinct())
DimDatesDF.count()
```

```python
#Rename the columns
DimDatesDF=DimDatesDF.withColumnRenamed('DISCOVERY_DATE_KEY','DATE_KEY')
DimDatesDF=DimDatesDF.withColumnRenamed('DISCOVERY_DATE','DATE')
DimDatesDF=DimDatesDF.withColumnRenamed('DISCOVERY_DOY','DAYOFYEAR')
```

# Create a Class for Database Connection

```python
class DBConn:
    def Conn(self,hostName,databseName,trusted):
        jdbcHostname = hostName
        jdbcDatabase = databseName
        jdbcUrl = "jdbc:postgresql://{0}:5432/{1}".format(jdbcHostname,jdbcDatabase)
        if (trusted):
            jdbcUrl = jdbcUrl+"integratedSecurity=True"
        return jdbcUrl
    def ConnProperties(self,user,password,trusted):
        connectionProperties={}
        if (trusted):
            connectionProperties = {
                "Driver" : "org.postgresql.Driver"
            }
        else:
            connectionProperties = {
                "Driver" : "org.postgresql.Driver",
                "user" : user,
                "password" : password
            }
        return connectionProperties
```

```python
#read config file to get POSTGRES user and password
config = configparser.ConfigParser()
config.read('config.cfg')
user=config['DB']['user']
password=config['DB']['password']
```

```python
#Instansiate the class
getConn=DBConn()
url=getConn.Conn("localhost","FIRES_DM",0)
properties=getConn.ConnProperties(user,password,0)
```

# Generic Load Data Function for all Data Frames

```python
def LoadData(df,url, table, mode, properties, output_data):
    """The function loads data to the input table and also creates a parquet in out folder"""
    df.write.parquet(output_data + table,  mode=mode)
    output=df.write.jdbc(url=url, table=table, mode=mode, properties=properties)
    return output
```
